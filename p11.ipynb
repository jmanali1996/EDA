## Converting SESSION NO to string data type

```python
df['SESSION NO'] = df['SESSION NO'].apply(str)
```

## Converting h:mm:ss time format to minutes

```python
df['SD'] = df['SESSION DURATION (h:mm:ss)'].str.split(':').apply(lambda x: int(x[0]) * 60 + int(x[1]) + int(x[2]) / 60)
df['AT'] = df['ACTIVE TIME (h:mm:ss)'].str.split(':').apply(lambda x: int(x[0]) * 60 + int(x[1]) + int(x[2]) / 60)
df['IT'] = df['IDLE TIME (h:mm:ss)'].str.split(':').apply(lambda x: int(x[0]) * 60 + int(x[1]) + int(x[2]) / 60)
df.head()
```

## Removing unecessary columns

```python
df = df.drop(['SESSION DURATION (h:mm:ss)', 'ACTIVE TIME (h:mm:ss)', 'IDLE TIME (h:mm:ss)'], axis = 1)
```

## Plot for SIDE variable

```python
# Plotting the count plot
plt.figure(figsize=(10, 6))
ax = sns.countplot(x = df['SIDE'], order = df['SIDE'].value_counts().index)
ax.bar_label(ax.containers[0], fmt = '%d', label_type = 'edge', fontsize = 10, padding = 3)

# Adding title and labels
plt.title('TIMES EACH SIDE WAS WORKED ON', fontsize = 16)
plt.xlabel('Side', fontsize = 12)
plt.ylabel('Count', fontsize = 12)
plt.tight_layout()

# Showing the plot
plt.show()
```

The above bar chart depicts which side of the body was worked out more. The right side worked out just two times more than the left side of the body.

## Plot for JOINT and SIDE variables

```python
# Creating a groupby dataframe JOINT counted by SIDE
sj = df.groupby('JOINT')['SIDE'].value_counts().reset_index(name = 'count')

# Plotting the bar chart
plt.figure(figsize = (10, 6))
ax = sns.barplot(data = sj, x = 'JOINT', y = 'count', hue = 'SIDE', palette = 'viridis')

# Adding bar labels
for container in ax.containers:
    ax.bar_label(container, fmt = '%d', label_type = 'edge', fontsize = 10, padding = 3)

# Adding title and labels
plt.title('WORKED OUT SIDE BY JOINT', fontsize = 16)
plt.xlabel('Joint', fontsize = 12)
plt.ylabel('Count', fontsize = 12)
plt.legend(title = 'Side')
plt.tight_layout()

# Showing the plot
plt.show()
```

The clustered bar chart reports which joint of the body worked out the most. The legend further indicates whether it was right joint or left joint. Both the knees worked out the maximum time followed by hips. Right shoulder and forearm worked equally but the left forearm worked three times lesser than the left shoulder. Both the wrists worked out once.

## Plot for EXERCISE variable

```python
# Creating data for pie chart
ex_counts = df['EXERCISE'].value_counts()
labels = ex_counts.index
sizes = ex_counts.values

# Custom function to show count and percentage in pie chart
def autopct_func(pct, all_vals):
    absolute = int(round(pct * sum(all_vals) / 100.0))  # Calculate absolute count
    return f"{pct:.1f}%\n({absolute})"  # Return percentage and count

palette = sns.color_palette('hls', len(labels))

# Plotting the pie chart
plt.figure(figsize = (8, 6))
plt.pie(
    sizes,
    labels = labels,
    autopct = lambda pct: autopct_func(pct, sizes),  # Custom autopct
    startangle = 140,
    colors = palette
)
plt.title("EXERCISES DISTRIBUTION")

# Showing the plot
plt.tight_layout()
plt.show()
```

The pie chart presents different types of the exercises, with their frequency and percentage, that the patient performed. Ext-flex sitting exercise was performed the most, followed by int-ext rotation, extension flexion, pron-sup spherical grip, and ext-flex w/ gravity.

## Plot for EXERCISE, SIDE and SETS variables

```python
# Creating a groupby dataframe of EXERCISE and SIDE, sumed up by SETS
es = df.groupby(['EXERCISE', 'SIDE'])['SETS'].sum().reset_index(name = 'count')

# Plotting the bar chart
plt.figure(figsize = (10, 6))
ax = sns.barplot(data = es, x='EXERCISE', y = 'count', hue = 'SIDE', palette = 'magma')

# Adding bar labels
for container in ax.containers:
    ax.bar_label(container, fmt = '%d', label_type = 'edge', fontsize = 10, padding = 3)

# Adding title and labels
plt.title("WORKED OUT SIDE BY EXERCISE", fontsize = 16)
plt.xlabel("Exercise", fontsize = 12)
plt.ylabel("Total Sets", fontsize = 12)
plt.legend(title = "Side")
plt.tight_layout()

# Showing the plot
plt.show()
```

This clustered bar chart shows the WORKED OUT SIDE BY EXERCISE. The patient performed five types of exercises - ext-flex sitting, ext-flex w/ gravity, extension flexion, int-ext rotation, and pron-sup spherical grip. The most sets were performed of the ext-flex sitting exercise followed by int-ext rotation while ext-flex w/ gravity was the exercise with the least number of sets performed for both the sides.

## Correlation between numerical variables

```python
# Calculating correlation and filtering for pairs with correlation > threshold
corr = df.select_dtypes(include=np.number).corr()
threshold = 0.7
high_corr_pairs = corr[(abs(corr) > threshold) & (corr != 1.0)].stack().index

# Number of scatter plots
num_plots = len(high_corr_pairs)

# Creating grid layout (2 columns, n rows)
n_cols = 2
n_rows = (num_plots + n_cols - 1) // n_cols  # Calculating required rows
fig, axes = plt.subplots(n_rows, n_cols, figsize=(16, n_rows * 4))

# Flattening axes array for easier indexing
axes = axes.flatten()

# Plotting the scatter plots
for i, (col1, col2) in enumerate(high_corr_pairs):
    sns.scatterplot(x = df[col1], y = df[col2], ax = axes[i], alpha = 0.7)
    axes[i].set_title(f"{col1} vs {col2} (Corr: {corr.loc[col1, col2]:.2f})")
    axes[i].set_xlabel(col1)
    axes[i].set_ylabel(col2)

# Adjusting layout and space between columns/rows
plt.subplots_adjust(hspace=0.5, wspace=0.5)

# Showing the plot
plt.show()
```

The above image are the correlation graphs between the variables SETS and TOTAL REPS, TOTAL REPS and CUMULATIVE MOVEMENT (deg), RANGE DURING ASSESSMENT (deg) and RANGE DURING SESSION (deg), SD and AT. Only pairs of thse variables were considered whose correlation value is greater than 7, with the exception of the variables correlating with itself that results in correlation value as 1.
1.   SETS and TOTAL REPS

*correlation value: 0.83*

With the increase in the number of TOTAL REPS, the number of SETS also increases, distributing the TOTAL REPS in an even manner, thus allowing the patient to exercise more efficiently.

2.   TOTAL REPS and CUMULATIVE MOVEMENT (deg)

*correlation value: 0.77*

More number of TOTAL REPS eventually results in higher degree of CUMULATIVE MOVEMENT. It's also evident that when a certain number of TOTAL REPS are performed in a discipline, CUMULATIVE MOVEMENT increases (30 -> 12000).

3.   RANGE DURING ASSESSMENT (deg) and RANGE DURING SESSION (deg)

*correlation value: 0.94*

The two variables increased in a linear fashion suggesting that the range that was noted before beginning the session - RANGE DURING ASSESSMENT - has only resulted in increament during the session. However, there's one outlier incident where the RANGE DURING ASSESSMENT was around 200 degrees and the RANGE DURING SESSION dropped to near 125 degrees.

4.   SD and AT

*correlation value: 0.99*

In every session, the active time was reasonably more. Where most sessions lasted for 20 minutes, only 3 sessions were longer than 35 minutes, the longest session being around 48 minutes.
